### 1.3 Exceptions

- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if type(a) != int and type(a) != float:
        raise TypeError("Inputs must be numeric")

    if type(b) != int and type(b) != float:
        raise TypeError("Inputs must be numeric")

    if b == 0:
        return None

    return a / b


# Tests
assert safe_divide(10,2) == 5.0
assert safe_divide(5,0) is None

try:
    safe_divide('x',2)
    assert False
except TypeError as e:
    assert str(e) == 'Inputs must be numeric'

print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# Plot eigenvectors scaled by eigenvalues
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    ax.quiver(
        0, 0,
        lam * v[0], lam * v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.axhline(0, color='red', lw=0.8)
ax.axvline(0, color='red', lw=0.8)
ax.legend()

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

** Geometric Explanation (4.3):**

>*Replace this text with your explanation of what eigenvalues/eigenvectors mean geometrically — what does it mean for a matrix to "stretch" or "squish" a vector?*

Eigenvectors are special directions that do not change direction when a matrix transformation is applied. They are only stretched, compressed, or flipped. 
The amount of stretching or shrinking is determined by the corresponding eigenvalue.
1) **eigenvalue > 1:** vector is stretched. 
2) **0 < eigenvalue < 1:**  vector is compressed. 
3) **eigenvalue < 0:**   the vector is flipped in the opposite direction as well as scaled. 
Geometrically, a matrix transformation can rotate and distort most vectors, but eigenvectors remain on the same line, making them important for understanding the behavior of the transformation.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

# SVD
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Reconstruct X
X_reconstructed = U @ np.diag(S) @ Vt

# Rank-1 approximation
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

>*Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*

Singular Value Decomposition (SVD) breaks a matrix into three parts: U, S, and Vt.
**- U:** information about the relationships between the data samples.       
**- S:** contains the singular values, which represent the importance or amount of information captured by each component.     
**- Vt:** contains the principal directions of the data.
In PCA, the rows of 'Vt'  represent the principal components. 
These are the directions along which the data varies the most. 
The singular values in 'S' tell us how much variance is captured by each principal component.
Therefore, the matrix 'Vt' contains the directions of maximum variance, while 'S' tells us the significance of those directions. By keeping only the top principal components, we can reduce dimensionality while preserving most of the important information in the dataset.

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.histogram_bin_edges(expected, bins=bins)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Avoid division by zero
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    psi = np.sum((actual_pct - expected_pct) *
                 np.log(actual_pct / expected_pct))

    return psi
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, alpha=0.6,color='pink' ,density=True, label='Training')
ax.hist(drift_dist, bins=30, alpha=0.6,color='green', density=True, label='Production')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** It is the Chanage in statistical properties of data overtime causing data used by Machine learning model in production different from data it was trained on. 
- **Covariate drift:** Distribution of input features changes over time
- **PSI < 0.1:** no change in both datasets
- **PSI 0.1–0.2:** Slight change occured
- **PSI > 0.2:** Large Shift in distribution has occured between both datasets
- **Retraining trigger example:**  If a fraud detection model was trained on last year's transaction data and the current production data shows a PSI greater than 0.2, it indicates a major shift in customer behavior. In such a case, retraining the model with recent data would be a good decision.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red/total
p_blue  = blue/total
p_green = green/total

# Joint: P(red first AND blue second)
p_red_then_blue = red/total * (blue/(total-1))

# Conditional: P(blue second | red first)
p_blue_given_red = blue/(total-1)

# Independence check: are P(B|A) == P(B) ?
independent = (p_blue_given_red == p_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam

    evidence = (p_word_given_spam * prior_spam) + \
               (p_word_given_ham * prior_ham)

    posterior = (p_word_given_spam * prior_spam) / evidence

    return posterior
    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
# Generate 5000 sample means
sample_means = np.array([
    np.mean(np.random.choice(population, size=sample_size))
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = population.std()
# CLT predicted standard deviation
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# Histogram of sample means
ax.hist(sample_means,bins=40,density=True,alpha=0.7,color='skyblue',label='Sample Means')

# Theoretical Normal Distribution
x = np.linspace(sample_means.min(), sample_means.max(), 300)

ax.plot(x,stats.norm.pdf(x, loc=pop_mean, scale=clt_std),color='red',linewidth=2,label='CLT Normal Approximation')

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

The Central Limit Theorem is important because it helps us understand the behavior of sample data. Even if the original data is not normally distributed, the average of many samples tends to form a normal distribution when the sample size is large enough.
This is useful in statistics and machine learning because many techniques like t-tests, confidence intervals, and hypothesis testing work better when data follows a normal distribution. CLT gives us confidence to use these methods even when the original data is not perfectly normal.
In this experiment, the original data followed an exponential distribution, but the distribution of the sample means became close to a normal distribution. We can also see that the actual standard deviation of the sample means (0.1828) is very close to the CLT predicted value (0.1827), which supports the Central Limit Theorem.
This is important because many statistical techniques such as confidence intervals, t-tests, and hypothesis testing assume that sample means are approximately normally distributed. CLT allows us to use these methods even when the original data is not normally distributed.
